```python
# Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?
```

```python
df.groupby("job_title").salary_in_usd.mean().max()
```

```python
df_groupby_job_title = df.groupby("job_title")

max_mean_salary = 0
max_mean_salary_group = ""

for group_name in df_groupby_job_title.groups.keys():
    if (
        df_groupby_job_title.get_group(group_name).salary_in_usd.mean()
        > max_mean_salary
    ):
        max_mean_salary = df_groupby_job_title.get_group(
            group_name
        ).salary_in_usd.mean()
        max_mean_salary_group = group_name

print("Max mean salary group:", max_mean_salary_group)
print("Max mean salary:", max_mean_salary)
```

```python
# Какое процентное соотношение каждого региона по вакансиям от всех вакансий?
# -> region_vacs_qty_percengage
```

```python
all_vacancies_count = df.count().max()
all_vacancies_count
```

```python
df_groupby_company_location = df.groupby("company_location")

df_groupby_company_location.job_title.count().keys()

region_vacs_qty_percengage = {}

for region_name in df_groupby_company_location.job_title.count().keys():
    region_vacs_qty_percengage[region_name] = round(
        int(df_groupby_company_location.job_title.count()[region_name])
        / all_vacancies_count
        * 100,
        2,
    )

region_vacs_qty_percengage
```

```python
# Какая корреляция уровня опыта от зарплаты?
# (Преобразовать уровень опыта в числовой тип, если возможно.
# Например: 'MI': 0, 'SE': 1, 'EN': 2, 'EX': 3)
# -> df_2.corr().loc["experience_level", "salary_in_usd"]
```

```python
df_2 = pd.DataFrame(df, columns=["experience_level", "salary_in_usd"])

df_2["experience_level"] = (
    df_2["experience_level"]
    .replace(["MI"], 0)
    .replace(["SE"], 1)
    .replace(["EN"], 2)
    .replace(["EX"], 3)
)

round(df_2.corr().loc["experience_level", "salary_in_usd"], 2)
```

```python
# Сколько должностей в наборе данных?
# -> len(vacancy_count.keys())
```

```python
df_groupby_job_title = df.groupby("job_title")

vacancy_count = {}

for job_title in df_groupby_job_title.job_title.count().keys():
    vacancy_count[job_title] = int(df_groupby_job_title.job_title.count()[job_title])

vacancy_count
len(vacancy_count.keys())
```

```python
# Какие 10 наиболее часто встречающихся должностей?
```

```python
for job_title in sorted(vacancy_count, key=vacancy_count.get, reverse=True)[:10]:
    print(f"{job_title}: {vacancy_count[job_title]}")
```
